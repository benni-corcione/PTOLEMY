```python
import sys
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
import math
```

### Funzioni di plot

```python
#funzione di plot
#plot di collection vs pinhole su geometrie diverse
def plot_geom (x1, y1, x2, y2, x3, y3):
    plt.figure(figsize=(8, 6))
    plt.plot(x1, y1, marker = 'o', markersize = 5, linewidth = 2, color = 'indianred', label='1 spacer up,   1 spacer down')
    plt.plot(x2, y2, marker = 'o', markersize = 5, linewidth = 2, color = 'mediumseagreen', label='2 spacers up, 1 spacer down')
    plt.plot(x3, y3, marker = 'o', markersize = 5, linewidth = 2, color = 'orange', label='1 spacer up,   2 spacers down')
    plt.xlabel('pinhole radius [$\mu$m]', size=15, loc='right')
    plt.ylabel('collection (%)', size=15, loc='top')
    plt.xticks(np.arange(0, 101, 10))
    #plt.xlim(0,32)
    plt.ylim(0,102)
    plt.xlim(0,101)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.grid("true")
    plt.legend(prop = { "size": 12 },framealpha=0.0)
    
    
#plot di collection vs pinhole su threshold diverse
def plot_thr (x1, y1, x2, y2):
    plt.figure(figsize=(8, 6))
    plt.plot(x1, y1, marker = 'o', markersize = 5, linewidth = 2, color = 'steelblue', label='V_threshold = - 100 V')
    plt.plot(x2, y2, marker = 'o', markersize = 5, linewidth = 2, color = 'orchid', label='V_threshold = - 50 V')
    plt.xlabel('pinhole radius [$\mu$m]', size=15, loc='right')
    plt.ylabel('collection (%)', size=15, loc='top')
    plt.xticks(np.arange(0, 101, 10))
    #plt.xlim(0,32)
    plt.ylim(0,102)
    plt.xlim(0,101)
    plt.xticks(fontsize=15)
    plt.yticks(fontsize=15)
    plt.grid("true")
    plt.legend(prop = { "size": 12 },framealpha=0.0)
```

### Collection vs pinhole (diverse geometrie)

```python
#array dati energy = 60 e soglia a 50 
#1 spacer up 1 spacer down
pin1 = [5, 10, 15, 25, 35, 50, 100]
f1 = [100, 100, 100, 61.1293, 30.1895, 14.1676, 3.51565]

#2 spacers up 1 spacer down
pin2 = [25, 35, 50, 100]
f2 = [68.1408, 32.6932, 16.2651, 4.12461]

#1 spacers up 2 spacers down
pin3 = [25]
f3 = [42.6009]

plot_geom(pin1,f1,pin2,f2,pin3,f3)
plt.axvline(30, color = 'black',linestyle='dashed', linewidth=2)
plt.title("$E_e$ = 60 eV, $V_{thr}$ = - 50 V")
plt.savefig("comparison_spacers.png")
```

### Collection vs pinhole (diverse threshold)

```python
#array dati vcnt = 50, geometria 1 spacer up 1 spacer down
#threshold -50 V
pin50_1 = [15, 25, 35, 50, 100]
f50_1 = [100, 66.7125, 34.4298, 17.9972, 4.18584]

#threshold -100 V
pin100_2 = [5, 10, 15, 25, 35, 50, 100]
f100_2 = [100, 100, 100, 54.7264, 28.5534, 14.5694, 3.54923] 


plot_thr(pin50_1,f50_1,pin100_2,f100_2)
plt.axvline(30, color = 'black',linestyle='dashed', linewidth=2)
plt.title("$E_e$ = 50 eV, 1 spacer up and 1 spacer down")
plt.savefig("comparison_thresholds.png")
```
