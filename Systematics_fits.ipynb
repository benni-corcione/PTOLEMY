```python
# Dati inseriti manualmente
Energy_CD204  = [96, 97, 98, 99, 100, 101, 102, 103]
Energyb_CD204  = [98,98,103,103]
syst_CB_gaus_CD204 = [10.3972, 16.8429, 6.8511, 4.39389, 13.6055, 10.3952, 14.794, 10.853 ]
syst_CB_gaus_CD204_err = [11.99, 10.33, 8.01, 6.17, 6.59, 9.41, 8.73, 7.02]
syst_R_gaus_CD204  = [0.1, 2, 2.3, 2.3]
syst_B_gaus_CD204  = [1.3, 2, 2.3, 1.3]

syst_CB_fwhm_CD204 = [9.2, 10.6, 5.7, 2.6, 14.3, 10.6, 8.8, 16]
syst_R_fwhm_CD204  = [0.1, 9.7, 12, 8.7]
syst_B_fwhm_CD204  = [0.5, 2.3, 4, 3]

media1 = sum(syst_CB_gaus_CD204) / len(syst_CB_gaus_CD204)
media2 = sum(syst_R_gaus_CD204) / len(syst_R_gaus_CD204)
media3 = sum(syst_B_gaus_CD204) / len(syst_B_gaus_CD204)

media4 = sum(syst_CB_fwhm_CD204) / len(syst_CB_fwhm_CD204)
media5 = sum(syst_R_fwhm_CD204) / len(syst_R_fwhm_CD204)
media6 = sum(syst_B_fwhm_CD204) / len(syst_B_fwhm_CD204)

print(media1)
print(media2)
print(media3)
print(media4)
print(media5)
print(media6)
```

```python
# Plot dei punti con plt.plot
plt.errorbar(Energy_CD204, syst_CB_gaus_CD204, yerr = syst_CB_gaus_CD204_err, marker='o', linestyle='None', color = 'tomato', label = "crystal ball")
plt.plot(Energyb_CD204, syst_R_gaus_CD204, marker='o', linestyle='None', color = 'cornflowerblue', label = "range")
plt.plot(Energyb_CD204, syst_B_gaus_CD204, marker='o', linestyle='None', color = 'gold', label = "binning")

# Etichette e titolo
plt.xlabel("Energy")
plt.ylabel("Discrepanza percentuale gaus")
plt.title("New dataset")
plt.ylim(-0.5,32)
plt.legend()

plt.savefig("syst_gaus_cd204.png", dpi=300)
plt.show()
```
